```python
gender=list(df['Sex'].unique())
embark = list(df['Embarked'].unique())
embark
```

```python
def gender_convert(x):
    
    return gender.index(x)

    
```

```python
df['Sex']=df['Sex'].apply(gender_convert)

df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode())

df['Age']=df['Age'].fillna(float(df['Age'].mode()))
df['Age']=df['Age'].astype(int)

df['Fare']=df['Fare'].fillna(float(df['Fare'].mode()))
df['Fare']=df['Fare'].astype(int)

df.head()
```
