```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('a10.csv')
df
```

```python
# features
df.info()
```

```python
for i in df.columns:
    print(i,"-->",df[i].dtypes)
```

```python
#histogram
features = list(df.columns)
features
```

```python
df.hist(column=features[1])
```

```python
plt.hist(df[features[1]])
plt.xlabel("length")
plt.ylabel("frequency")
plt.title("sepal lenth cm")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(20,20))
for i in range(1,len(features)):
    plt.subplot(3,2,i)
    plt.hist(df[features[i]])
    plt.xlabel(features[i])
    plt.ylabel('frequency')
    plt.title(features[i])

plt.show()
```

```python

plt.figure(figsize=(20,20))
for i in range(1,len(features)-1):
    plt.subplot(3,2,i)
    plt.boxplot(df[features[i]])
    plt.xlabel(features[i])
plt.show()
```
