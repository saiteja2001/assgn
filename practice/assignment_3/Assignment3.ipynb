```python
grp_species = df_iris.groupby('Species')
grp_species.first()
```

```python
grp_species.describe()
```

```python
# min values
grp_species.agg(lambda x: x.min())[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
```

```python
# max values
grp_species.agg(lambda x: x.max())[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
```

```python
# mean
grp_species.agg(lambda x: x.mean())[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
```

```python
# median
grp_species.agg(lambda x: x.median())[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
```

```python
# variance
grp_species.agg(lambda x: x.var())[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
```

```python
# standard deviation
grp_species.agg(lambda x: x.std())[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
```

```python
# mean absolute deviation
grp_species.agg(lambda x: x.mad())[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
```
